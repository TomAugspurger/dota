We can estimate the Gini coefficient $G$ by
$G = \frac{\sum_{i=1}^n (2i - n - 1)x_i^T}{n^2 \mu}$.

```python
def gini(team):
    n = team.shape[0]
    num = np.arange(n, 0, -1).dot(team.net_worth.values)
    denom = team.net_worth.sum()
    g = n + 1 - 2 * (num / denom) / n
    return g
```
