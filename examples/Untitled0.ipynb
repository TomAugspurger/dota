```python
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker

from sqlalchemyORM import Game, Player, PlayerGame, Team

from dota import api
from dota.api import _hero_id_to_names
```

```python
# get from dota.api._680
_680 = 1390802400
```

```python
engine_path = Path("../data/pro/pro.db").absolute()
engine = create_engine("sqlite:///" + str(engine_path))
Session = sessionmaker(bind=engine)
session = Session()
```

```python
games = session.query(Game).filter(Game.start_time > _680)
print(len(games.all()))
```

```python
# example for one
g = games.first()
pgs = session.query(PlayerGame).filter(PlayerGame.match_id == g.match_id).all()

keys = ['level', 'kills', 'deaths', 'assists',
        'last_hits', 'denies', 'gold', 'gold_spent', 'player_slot',
        'account_id', 'hero_damage', 'hero_healing',
        'item_0', 'item_1', 'item_2', 'item_3', 'item_4',
        'item_5', 'hero_id']

data = [{k: getattr(pg, k) for k in keys} for pg in pgs]
df = pd.DataFrame(data)

wd = {True: "Radiant", False: "Dire"}

df = api.DetailsResponse.format_df(df, wd[g.radiant_win], g.match_id)
df.head()
```

```python
keys = ['level', 'kills', 'deaths', 'assists',
        'last_hits', 'denies', 'gold', 'gold_spent', 'player_slot',
        'account_id', 'hero_damage', 'hero_healing',
        'item_0', 'item_1', 'item_2', 'item_3', 'item_4',
        'item_5', 'hero_id']
wd = {True: "Radiant", False: "Dire"}

dfs = []
for g in games:
    pgs = session.query(PlayerGame).filter(PlayerGame.match_id == g.match_id).all()

    data = [{k: getattr(pg, k) for k in keys} for pg in pgs]
    df = pd.DataFrame(data)

    df = api.DetailsResponse.format_df(df, wd[g.radiant_win], g.match_id)
    dfs.append(df)
    
df = pd.concat(dfs)
```
