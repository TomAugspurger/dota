```python
%matplotlib inline

import os
import pathlib

import arrow
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from dota.helpers import cached_games
from dota import api
```

```python
data_dir = pathlib.Path(os.path.expanduser('~/sandbox/dota/data/pro'))
games = cached_games(data_dir)

# Limit to 6.80 and filter modes with picks or bans
_680 = arrow.get(1381989600)

pbs = (api.DetailsResponse.from_json(str(data_dir) + '/' + str(x) + '.json') for x in games)
pbs = (dr for dr in pbs if getattr(dr, 'picks_bans') is not None
       and dr.start_time > _680)
df = pd.concat(x.picks_bans for x in pbs)
df['hero_id'] = df.hero_id.replace(api._hero_id_to_names)
df.head(20)
```

```python
picks = df[df.is_pick]['hero_id'].value_counts()
bans = df[~df.is_pick]['hero_id'].value_counts()
either = (picks + bans).dropna()
either.sort()

# resort for plot
picks = picks.reindex_like(either)
bans = bans.reindex_like(either)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(20, 50), ncols=3)
either.plot(kind='barh', ax=ax0)
ax0.set_title("Picked or Banned")

picks.plot(kind='barh', ax=ax1)
ax1.set_title("Picked")

bans.plot(kind='barh', ax=ax2)
ax2.set_title("Banned")
```

```python
# Include the time
data_dir = pathlib.Path(os.path.expanduser('~/sandbox/dota/data/pro'))
games = cached_games(data_dir)

# Limit to 6.80 and filter modes with picks or bans
_680 = arrow.get(1381989600)

pbs = (api.DetailsResponse.from_json(str(data_dir) + '/' + str(x) + '.json') for x in games)
pbs = filter(lambda dr: (getattr(dr, 'picks_bans') is not None
                         and dr.start_time > _680), pbs)
dfs = []
for pb in pbs:
    df = pb.picks_bans
    df['start_time'] = pb.start_time.isoformat()
    dfs.append(df)

df = pd.concat(dfs, ignore_index=True)
df['hero_id'] = df.hero_id.replace(api._hero_id_to_names)
df['start_time'] = pd.to_datetime(df.start_time)
df.head(20)
```

```python
# pd.Grouper available from pandas >= .14

g = df.groupby([pd.Grouper(freq='7d', key='start_time'), 'hero_id'])['hero_id']
by_time = g.count()
# normalize (20 picks per game)
by_time = by_time.div(by_time.sum(level='start_time') / 20)
```

```python
by_time.loc[:, 'invoker'].plot(label='invoker')  # right censored
by_time.loc[:, 'elder_titan'].plot(label='ET')  # right censored
by_time.loc[:, 'shredder'].plot(label='timber')  # right censored
plt.legend(loc=2)
```
